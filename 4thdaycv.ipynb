````python
## 📥 Step 2: Import Libraries
# cv2        → OpenCV for image processing and text drawing
# numpy      → for creating blank canvas (array)
# matplotlib → for displaying images inside notebook

import cv2
import numpy as np
import matplotlib.pyplot as plt

print("✅ All Libraries Imported Successfully!")
````

````python
## 🛠️ Step 3: Create Helper Function
# We create a reusable function to display images
#So we don't repeat the same display code every time
def show(image, title='Image', size=(7, 7)):
    """
    Helper function to display image in notebook.
    
    image → the image to display
    title → title shown above image
    size  → figure size (width, height)
    """
    plt.figure(figsize=size)
    plt.imshow(image)
    plt.title(title, fontsize=14, fontweight='bold')
    plt.axis('off')   # hide axis numbers
    plt.tight_layout()
    plt.show()


def blank_canvas(height=400, width=600, color='white'):
    """
    Create a blank canvas.
    
    color = 'white' → white background (255,255,255)
    color = 'black' → black background (0,0,0)
    """
    canvas = np.ones((height, width, 3), dtype=np.uint8)

    if color == 'white':
        canvas *= 255    # fill with white
    else:
        canvas *= 0      # fill with black

    return canvas

print("✅ Helper Functions Ready!")
````

# 🟢 LEVEL 1 — BEGINNER
## 📝 Part 1: Basic Text

**Syntax:**
```python
cv2.putText(image, text, position, font, fontScale, color, thickness)
Parameter
Meaning
image
Canvas to draw on
text
The string to write
position
(x, y) — bottom-left of text
font
Font style (cv2.FONT_...)
fontScale
Size of text (1.0 = normal)
color
(B, G, R) color
thickness
Stroke thickness in pixels

````python
# ── Create white canvas ──
canvas = blank_canvas(400, 600, 'white')

# ─────────────────────────────────────────
# cv2.putText() — most basic text function
#
# Parameters:
# image     → where to draw
# text      → what to write
# org       → (x, y) bottom-left position
# fontFace  → font style
# fontScale → text size multiplier
# color     → (B, G, R)
# thickness → line thickness
# ─────────────────────────────────────────

# Simple black text
cv2.putText(
    canvas,                          # image/canvas
    'Hello, OpenCV!',                # text to write
    (50, 80),                        # position (x=50, y=80)
    cv2.FONT_HERSHEY_SIMPLEX,        # font style
    1.5,                             # font size (1.5x normal)
    (0, 0, 0),                       # color → BLACK (B,G,R)
    2                                # thickness = 2px
)

# Simple red text
cv2.putText(
    canvas,
    'Learning OpenCV!',
    (50, 160),                       # move down to y=160
    cv2.FONT_HERSHEY_SIMPLEX,
    1.2,
    (0, 0, 255),                     # RED in BGR
    2
)

# Simple blue text
cv2.putText(
    canvas,
    'Text is Easy!',
    (50, 240),
    cv2.FONT_HERSHEY_SIMPLEX,
    1.2,
    (255, 0, 0),                     # BLUE in BGR
    2
)

# Simple green text
cv2.putText(
    canvas,
    'Keep Practicing!',
    (50, 320),
    cv2.FONT_HERSHEY_SIMPLEX,
    1.2,
    (0, 180, 0),                     # GREEN in BGR
    2
)

show(canvas, '📝 Basic Text Examples')
````

## 🔤 Part 2: All Font Styles in OpenCV

OpenCV has **8 built-in font styles**:

| Font Constant | Style |
|--------------|-------|
| `FONT_HERSHEY_SIMPLEX` | Normal sans-serif |
| `FONT_HERSHEY_PLAIN` | Small plain font |
| `FONT_HERSHEY_DUPLEX` | Thicker sans-serif |
| `FONT_HERSHEY_COMPLEX` | Serif font |
| `FONT_HERSHEY_TRIPLEX` | Thick serif |
| `FONT_HERSHEY_COMPLEX_SMALL` | Small serif |
| `FONT_HERSHEY_SCRIPT_SIMPLEX` | Handwriting style |
| `FONT_HERSHEY_SCRIPT_COMPLEX` | Thick handwriting |
| `+ FONT_ITALIC` | Add italic to any font |

````python
# ── Create white canvas ──
canvas = blank_canvas(520, 700, 'white')

# ─────────────────────────────────────────
# All 8 OpenCV font styles
# We loop through each and display them
# ─────────────────────────────────────────

# List of all fonts with their names
fonts = [
    (cv2.FONT_HERSHEY_SIMPLEX,        'SIMPLEX'),
    (cv2.FONT_HERSHEY_PLAIN,          'PLAIN'),
    (cv2.FONT_HERSHEY_DUPLEX,         'DUPLEX'),
    (cv2.FONT_HERSHEY_COMPLEX,        'COMPLEX'),
    (cv2.FONT_HERSHEY_TRIPLEX,        'TRIPLEX'),
    (cv2.FONT_HERSHEY_COMPLEX_SMALL,  'COMPLEX SMALL'),
    (cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 'SCRIPT SIMPLEX'),
    (cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 'SCRIPT COMPLEX'),
]

# Starting y position
y = 55

for font, name in fonts:
    # Draw font name label (small gray text)
    cv2.putText(canvas, f'{name}:',
                (20, y - 5),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.4, (150, 150, 150), 1)

    # Draw sample text in that font
    cv2.putText(canvas,
                'Hello OpenCV 123',   # sample text
                (20, y + 25),
                font,                  # current font
                0.9,                   # size
                (30, 30, 30),          # dark gray color
                1)                     # thickness

    y += 60   # move down for next font

show(canvas, '🔤 All Font Styles', size=(9, 8))
````

## 📐 Part 3: Text Size & Thickness Variations

````python
# ── Create white canvas ──
canvas = blank_canvas(500, 700, 'white')

# ─────────────────────────────────────────
# fontScale controls SIZE of text
# thickness controls BOLDNESS of text
# ─────────────────────────────────────────

# Different sizes (fontScale)
sizes = [0.5, 0.8, 1.0, 1.5, 2.0]
y = 60
cv2.putText(canvas, 'SIZE VARIATIONS:',
            (20, 30), cv2.FONT_HERSHEY_SIMPLEX,
            0.6, (100, 100, 100), 1)
for size in sizes:
    cv2.putText(canvas,
                f'Scale = {size}',
                (20, y),
                cv2.FONT_HERSHEY_SIMPLEX,
                size,              # changing size
                (0, 0, 200),       # blue
                2)
    y += int(size * 50 + 20)   # spacing based on size

# Different thickness
cv2.putText(canvas, 'THICKNESS VARIATIONS:',
            (380, 30), cv2.FONT_HERSHEY_SIMPLEX,
            0.6, (100, 100, 100), 1)

thicknesses = [1, 2, 3, 5, 8]
y2 = 80

for thick in thicknesses:
    cv2.putText(canvas,
                f'Thick={thick}',
                (380, y2),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.9,
                (200, 0, 0),       # red
                thick)             # changing thickness
    y2 += 70

show(canvas, '📐 Size & Thickness Variations', size=(10, 7))
````

## 📍 Part 4: Text Positioning on Canvas

> Learn how to place text at:
> - Corners (Top-Left, Top-Right, Bottom-Left, Bottom-Right)
> - Center of image
> - Custom position

> 💡 Use `cv2.getTextSize()` to measure text width/height
> before placing it — very useful for centering!

````python
# ── Create white canvas ──
canvas = blank_canvas(500, 700, 'white')

# Draw border for reference
cv2.rectangle(canvas, (10, 10), (690, 490), (200, 200, 200), 2)

# ─────────────────────────────────────────
# cv2.getTextSize() → returns (width, height) of text
# Very useful to calculate center position!
#
# Usage:
# (text_w, text_h), baseline = cv2.getTextSize(
#     text, font, fontScale, thickness)
# ─────────────────────────────────────────

font       = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.8
thickness  = 2
color      = (50, 50, 200)

# ── TOP LEFT ──
cv2.putText(canvas, 'Top Left',
            (20, 40),   # small x, small y
            font, font_scale, color, thickness)

# ── TOP RIGHT ──
text = 'Top Right'
# Measure text size first
(tw, th), _ = cv2.getTextSize(text, font, font_scale, thickness)
# x = canvas_width - text_width - margin
cv2.putText(canvas, text,
            (700 - tw - 20, 40),
            font, font_scale, color, thickness)

# ── BOTTOM LEFT ──
cv2.putText(canvas, 'Bottom Left',
            (20, 475),   # small x, large y
            font, font_scale, color, thickness)

# ── BOTTOM RIGHT ──
text = 'Bottom Right'
(tw, th), _ = cv2.getTextSize(text, font, font_scale, thickness)
cv2.putText(canvas, text,
            (700 - tw - 20, 475),
            font, font_scale, color, thickness)

# ── CENTER ──
text = '★ CENTER ★'
(tw, th), _ = cv2.getTextSize(text, font, 1.2, 2)
# x = (canvas_width  - text_width)  // 2
# y = (canvas_height + text_height) // 2
cx = (700 - tw) // 2
cy = (500 + th) // 2
cv2.putText(canvas, text,
            (cx, cy),
            font, 1.2, (0, 150, 0), 2)

# Draw center dot for reference
cv2.circle(canvas, (350, 250), 5, (255, 0, 0), -1)

show(canvas, '📍 Text Positioning', size=(10, 7))
````

## 🔲 Part 5: Text with Background Box

> Adding a **colored box behind text** makes it
> readable on any background image!

**Steps:**
1. Measure text size using `cv2.getTextSize()`
2. Draw a filled rectangle behind the text
3. Draw the text on top

````python
# ── Create black canvas ──
canvas = blank_canvas(400, 700, 'black')

# ─────────────────────────────────────────
# Function to draw text with background box
# This is very useful for labels on images!
# ─────────────────────────────────────────

def draw_text_with_box(img, text, position,
                        font=cv2.FONT_HERSHEY_SIMPLEX,
                        font_scale=1.0,
                        text_color=(255, 255, 255),
                        box_color=(0, 0, 255),
                        thickness=2,
                        padding=10):
    """
    Draw text with a colored background box.

    img        → image to draw on
    text       → text string
    position   → (x, y) bottom-left of text
    text_color → color of text (B,G,R)
    box_color  → color of background box (B,G,R)
    padding    → space around text inside box
    """
    x, y = position

    # Step 1: Measure the text size
    (text_w, text_h), baseline = cv2.getTextSize(
        text, font, font_scale, thickness)

    # Step 2: Calculate box coordinates
    # Box starts above the text (y - text_h - padding)
    box_x1 = x - padding
    box_y1 = y - text_h - padding
    box_x2 = x + text_w + padding
    box_y2 = y + baseline + padding

    # Step 3: Draw filled rectangle (background box)
    cv2.rectangle(img,
                  (box_x1, box_y1),
                  (box_x2, box_y2),
                  box_color,
                  -1)   # -1 = filled

    # Step 4: Draw text on top of box
    cv2.putText(img, text, (x, y),
                font, font_scale,
                text_color, thickness)


# ── Use the function with different colors ──

draw_text_with_box(canvas, 'Red Box Label',
                   (50, 80),
                   box_color=(0, 0, 200),
                   text_color=(255, 255, 255))

draw_text_with_box(canvas, 'Green Box Label',
                   (50, 180),
                   box_color=(0, 180, 0),
                   text_color=(255, 255, 255))

draw_text_with_box(canvas, 'Blue Box Label',
                   (50, 280),
                   box_color=(200, 0, 0),
                   text_color=(255, 255, 0))

draw_text_with_box(canvas, 'Custom Padding',
                   (350, 180),
                   box_color=(128, 0, 128),
                   text_color=(255, 255, 255),
                   padding=20)           # bigger padding

draw_text_with_box(canvas, 'Small Text',
                   (350, 300),
                   font_scale=0.7,
                   box_color=(0, 165, 255),
                   text_color=(0, 0, 0))

show(canvas, '🔲 Text with Background Box', size=(10, 6))
````

## 💫 Part 6: Shadow & Outline Text Effect

> **Shadow** = draw dark text slightly offset, then draw colored text on top
> **Outline** = draw thick dark text first, then thin colored text on top

````python
# ── Create white canvas ──
canvas = blank_canvas(400, 700, 'white')

font = cv2.FONT_HERSHEY_SIMPLEX

# ─────────────────────────────────────────
# SHADOW TEXT EFFECT
# Step 1: Draw dark/gray text at offset position (x+3, y+3)
# Step 2: Draw main colored text at original position (x, y)
# The offset creates a shadow illusion!
# ─────────────────────────────────────────

text   = 'Shadow Effect!'
x, y   = 50, 100

# Step 1: Draw shadow (dark gray, offset by 4px)
cv2.putText(canvas, text,
            (x + 4, y + 4),      # offset position = shadow
            font, 1.5,
            (150, 150, 150),     # gray shadow color
            3)

# Step 2: Draw main text on top
cv2.putText(canvas, text,
            (x, y),              # original position
            font, 1.5,
            (0, 0, 200),         # blue main text
            3)

# ─────────────────────────────────────────
# OUTLINE TEXT EFFECT
# Step 1: Draw THICK text in dark color (outline)
# Step 2: Draw THIN text in light color on top (fill)
# ─────────────────────────────────────────

text2  = 'Outline Effect!'
x2, y2 = 50, 220

# Step 1: Draw thick outline (black, thickness=6)
cv2.putText(canvas, text2,
            (x2, y2),
            font, 1.5,
            (0, 0, 0),           # black outline
            6)                   # thick = outline

# Step 2: Draw thin fill on top (white, thickness=2)
cv2.putText(canvas, text2,
            (x2, y2),
            font, 1.5,
            (255, 255, 255),     # white fill
            2)                   # thin = fill

# ─────────────────────────────────────────
# COMBINED: Shadow + Outline
# ─────────────────────────────────────────

text3  = 'Shadow + Outline!'
x3, y3 = 50, 340

# Shadow layer
cv2.putText(canvas, text3, (x3+4, y3+4), font, 1.5, (100,100,100), 6)
# Outline layer
cv2.putText(canvas, text3, (x3, y3),     font, 1.5, (0, 0, 0),     6)
# Fill layer
cv2.putText(canvas, text3, (x3, y3),     font, 1.5, (0, 200, 255), 2)

show(canvas, '💫 Shadow & Outline Text Effects', size=(10, 6))
````

## 🌈 Part 7: Multiline Text

> OpenCV does **NOT** support `\n` newlines automatically
> We must manually calculate Y position for each line!

````python
# ── Create white canvas ──
canvas = blank_canvas(500, 700, 'white')

# ─────────────────────────────────────────
# Function to draw multiline text
# Splits text by '\n' and draws each line
# with proper spacing
# ─────────────────────────────────────────

def draw_multiline_text(img, text, position,
                         font=cv2.FONT_HERSHEY_SIMPLEX,
                         font_scale=0.8,
                         color=(0, 0, 0),
                         thickness=2,
                         line_spacing=15):
    """
    Draw text with multiple lines.

    text         → use '\n' to separate lines
    line_spacing → extra pixels between lines
    """
    x, y = position

    # Split text into individual lines
    lines = text.split('\n')

    for line in lines:
        # Draw this line
        cv2.putText(img, line, (x, y),
                    font, font_scale, color, thickness)

        # Measure line height to move down
        (_, line_h), _ = cv2.getTextSize(
            line, font, font_scale, thickness)

        # Move y down by line height + spacing
        y += line_h + line_spacing


# ── Example 1: Simple multiline ──
draw_multiline_text(
    canvas,
    'Line 1: Hello!\nLine 2: OpenCV\nLine 3: Is Fun!\nLine 4: Keep Going!',
    (50, 60),
    font_scale=1.0,
    color=(0, 0, 180),
    line_spacing=20
)

# ── Example 2: Paragraph style ──
paragraph = (
    'OpenCV is an open-source\n'
    'computer vision library.\n'
    'It helps us process images\n'
    'and videos easily in Python!'
)

draw_multiline_text(
    canvas,
    paragraph,
    (350, 60),
    font_scale=0.7,
    color=(0, 120, 0),
    line_spacing=12
)

# ── Example 3: Mixed colors per line ──
lines_colors = [
    ('RED   line text',   (0, 0, 255)),
    ('GREEN line text',   (0, 180, 0)),
    ('BLUE  line text',   (255, 0, 0)),
    ('BLACK line text',   (0, 0, 0)),
]

y_pos = 300
for line, clr in lines_colors:
    cv2.putText(canvas, line, (50, y_pos),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.9, clr, 2)
    y_pos += 50

show(canvas, '🌈 Multiline Text', size=(10, 7))
````
