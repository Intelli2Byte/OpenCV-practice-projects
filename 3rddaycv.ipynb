````python
# ─────────────────────────────────────────
# np.ones() creates an array filled with 1s
# Multiply by 255 → makes it WHITE
# Shape → (height, width, 3 color channels)
# dtype=np.uint8 → standard image data type (0-255)
# ─────────────────────────────────────────

# Create a white canvas of size 500 x 500
canvas = np.ones((500, 500, 3), dtype=np.uint8) * 255

print(f"✅ Canvas Created!")
print(f"📐 Canvas Shape : {canvas.shape}")
print(f"   Height : 500px")
print(f"   Width  : 500px")
print(f"   Color  : White (255, 255, 255)")

# Display the blank canvas
plt.figure(figsize=(5, 5))
plt.imshow(canvas)
plt.title('Blank White Canvas', fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
````

## 📏 Part 1: Drawing Lines

**Syntax:**
```python
cv2.line(image, start_point, end_point, color, thickness)


````python
# Create a fresh white canvas for lines
canvas_lines = np.ones((500, 500, 3), dtype=np.uint8) * 255

# ─────────────────────────────────────────
# cv2.line(canvas, start, end, color, thickness)
# Points are (x, y) → (column, row)
# ─────────────────────────────────────────

# 1. Simple Horizontal Line (Red)
# Starts at left (50, 80) → ends at right (450, 80)
cv2.line(canvas_lines,
         (50, 80),       # start point (x, y)
         (450, 80),      # end point   (x, y)
         (0, 0, 255),    # color → RED in BGR
         3)              # thickness = 3 pixels

# 2. Vertical Line (Blue)
# Starts at top (250, 50) → ends at bottom (250, 200)
cv2.line(canvas_lines,
         (250, 50),      # start point
         (250, 200),     # end point
         (255, 0, 0),    # color → BLUE in BGR
         3)              # thickness

# 3. Diagonal Line (Green)
# Goes from top-left to bottom-right
cv2.line(canvas_lines,
         (50, 150),      # start point
         (450, 350),     # end point
         (0, 255, 0),    # color → GREEN in BGR
         3)              # thickness

# 4. Thick Dashed-look Line (Purple)
cv2.line(canvas_lines,
         (50, 420),
         (450, 420),
         (128, 0, 128),  # color → PURPLE in BGR
         8)              # thick line = 8 pixels

# 5. Diagonal Line other direction (Orange)
cv2.line(canvas_lines,
         (450, 150),
         (50, 350),
         (0, 165, 255),  # color → ORANGE in BGR
         3)

# Add text labels on canvas
cv2.putText(canvas_lines, 'Horizontal (Red)',
            (55, 70), cv2.FONT_HERSHEY_SIMPLEX,
            0.5, (0, 0, 255), 1)

cv2.putText(canvas_lines, 'Vertical (Blue)',
            (255, 45), cv2.FONT_HERSHEY_SIMPLEX,
            0.5, (255, 0, 0), 1)

cv2.putText(canvas_lines, 'Diagonal (Green)',
            (55, 145), cv2.FONT_HERSHEY_SIMPLEX,
            0.5, (0, 200, 0), 1)

cv2.putText(canvas_lines, 'Thick Line (Purple)',
            (55, 415), cv2.FONT_HERSHEY_SIMPLEX,
            0.5, (128, 0, 128), 1)

# Display
plt.figure(figsize=(7, 7))
plt.imshow(canvas_lines)
plt.title('📏 Drawing Lines', fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
````

## ⬜ Part 2: Rectangle / Box

**Syntax:**
```python
cv2.rectangle(image, top_left, bottom_right, color, thickness)
Parameter
Meaning
top_left
(x, y) top-left corner
bottom_right
(x, y) bottom-right corner
thickness
Use -1 to fill the shape

````python
## ⬜ Cell 8 — Rectangle Code
# Create a fresh white canvas
canvas_rect = np.ones((500, 500, 3), dtype=np.uint8) * 255

# ─────────────────────────────────────────
# cv2.rectangle(canvas, top_left, bottom_right, color, thickness)
# thickness = -1 → fills the rectangle with color
# ─────────────────────────────────────────

# 1. Simple Rectangle (outline only) — Blue
cv2.rectangle(canvas_rect,
              (50, 50),       # top-left corner
              (200, 150),     # bottom-right corner
              (255, 0, 0),    # color → BLUE
              3)              # thickness = 3 (outline only)

# 2. Filled Rectangle — Green
cv2.rectangle(canvas_rect,
              (250, 50),      # top-left corner
              (450, 150),     # bottom-right corner
              (0, 200, 0),    # color → GREEN
              -1)             # -1 = FILLED (solid color)

# 3. Thick Border Rectangle — Red
cv2.rectangle(canvas_rect,
              (50, 200),
              (200, 320),
              (0, 0, 255),    # RED
              8)              # thick border = 8px

# 4. Square — Orange (equal width & height)
cv2.rectangle(canvas_rect,
              (250, 200),
              (430, 380),     # same width & height = square
              (0, 165, 255),  # ORANGE
              4)

# 5. Small filled box — Purple
cv2.rectangle(canvas_rect,
              (150, 400),
              (350, 470),
              (128, 0, 128),  # PURPLE
              -1)             # filled

# Add labels
cv2.putText(canvas_rect, 'Outline',   (55,  45), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 1)
cv2.putText(canvas_rect, 'Filled',    (255, 45), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,150,0), 1)
cv2.putText(canvas_rect, 'Thick',     (55, 195), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)
cv2.putText(canvas_rect, 'Square',    (255,195), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,130,200),1)
cv2.putText(canvas_rect, 'Filled Box',(155,395), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255),1)

# Display
plt.figure(figsize=(7, 7))
plt.imshow(canvas_rect)
plt.title('⬜ Drawing Rectangles & Boxes', fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
````

## ⭕ Part 3: Circle

**Syntax:**
```python
cv2.circle(image, center, radius, color, thickness)
Parameter
Meaning
center
(x, y) center point of circle
radius
radius in pixels
thickness
-1 = filled circle

````python
## ⭕ Cell 10 — Circle Code
# Create a fresh white canvas
canvas_circle = np.ones((500, 500, 3), dtype=np.uint8) * 255

# ─────────────────────────────────────────
# cv2.circle(canvas, center, radius, color, thickness)
# ─────────────────────────────────────────

# 1. Simple circle outline — Blue
cv2.circle(canvas_circle,
           (120, 120),    # center point (x, y)
           80,            # radius = 80 pixels
           (255, 0, 0),   # BLUE
           3)             # outline only

# 2. Filled circle — Red
cv2.circle(canvas_circle,
           (350, 120),    # center
           80,            # radius
           (0, 0, 255),   # RED
           -1)            # -1 = filled

# 3. Large circle — Green outline
cv2.circle(canvas_circle,
           (250, 320),    # center
           120,           # bigger radius
           (0, 200, 0),   # GREEN
           5)             # thick outline

# 4. Small filled dot — Orange
cv2.circle(canvas_circle,
           (250, 320),    # same center as above
           10,            # tiny radius = dot
           (0, 165, 255), # ORANGE
           -1)            # filled dot

# Add labels
cv2.putText(canvas_circle, 'Outline', (75,  40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 1)
cv2.putText(canvas_circle, 'Filled',  (310, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,200), 1)
cv2.putText(canvas_circle, 'Large + Center Dot', (130, 200), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,150,0), 1)

# Display
plt.figure(figsize=(7, 7))
plt.imshow(canvas_circle)
plt.title('⭕ Drawing Circles', fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
````

## 🔺 Part 4: Triangle & Polygon

**Syntax:**
```python
cv2.polylines(image, [points], isClosed, color, thickness)
cv2.fillPoly(image, [points], color)   # for filled shape

````python
# Create a fresh white canvas
canvas_poly = np.ones((500, 500, 3), dtype=np.uint8) * 255

# ─────────────────────────────────────────
# TRIANGLE
# Define 3 corner points (x, y)
# np.array → stores the points
# dtype=np.int32 → integer coordinates
# ─────────────────────────────────────────

# Triangle points (3 corners)
triangle_points = np.array([
    [150, 50],    # top point
    [50,  200],   # bottom-left point
    [250, 200]    # bottom-right point
], dtype=np.int32)

# Draw triangle outline — Blue
cv2.polylines(canvas_poly,
              [triangle_points],  # list of point arrays
              isClosed=True,      # True = connect last point to first
              color=(255, 0, 0),  # BLUE
              thickness=3)

# ─────────────────────────────────────────
# FILLED TRIANGLE — Green
# ─────────────────────────────────────────
filled_triangle = np.array([
    [350, 50],
    [270, 200],
    [450, 200]
], dtype=np.int32)

# fillPoly fills the shape with solid color
cv2.fillPoly(canvas_poly,
             [filled_triangle],   # list of point arrays
             color=(0, 200, 0))   # GREEN filled

# ─────────────────────────────────────────
# PENTAGON (5 sides)
# ─────────────────────────────────────────
# Calculate pentagon points using math
import math

# Center and radius of pentagon
cx, cy, r = 150, 370, 100

# Generate 5 equally spaced points around a circle
pentagon_points = np.array([
    [int(cx + r * math.cos(math.radians(90 + i * 72))),
     int(cy - r * math.sin(math.radians(90 + i * 72)))]
    for i in range(5)
], dtype=np.int32)

cv2.polylines(canvas_poly,
              [pentagon_points],
              isClosed=True,
              color=(0, 0, 255),   # RED
              thickness=3)

# ─────────────────────────────────────────
# HEXAGON (6 sides) — Filled Orange
# ─────────────────────────────────────────
cx2, cy2, r2 = 370, 370, 90

hexagon_points = np.array([
    [int(cx2 + r2 * math.cos(math.radians(i * 60))),
     int(cy2 + r2 * math.sin(math.radians(i * 60)))]
    for i in range(6)
], dtype=np.int32)

cv2.fillPoly(canvas_poly,
             [hexagon_points],
             color=(0, 165, 255))  # ORANGE filled

# Add labels
cv2.putText(canvas_poly, 'Triangle',        (80,  245), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0),   1)
cv2.putText(canvas_poly, 'Filled Triangle', (275, 245), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,150,0),   1)
cv2.putText(canvas_poly, 'Pentagon',        (90,  480), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,200),   1)
cv2.putText(canvas_poly, 'Hexagon',         (320, 480), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,130,200), 1)

# Display
plt.figure(figsize=(7, 7))
plt.imshow(canvas_poly)
plt.title('🔺 Triangle & Polygons', fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
````

## ❤️ Part 5: Heart Shape

> A heart is drawn using **bezier curve points**
> We calculate the heart coordinates using a **math formula**
> Then connect all points using `cv2.fillPoly()`

````python
# Create a fresh white canvas
canvas_heart = np.ones((500, 500, 3), dtype=np.uint8) * 255

# ─────────────────────────────────────────
# HEART SHAPE using parametric equations:
#
# x = 16 * sin³(t)
# y = 13cos(t) - 5cos(2t) - 2cos(3t) - cos(4t)
#
# t goes from 0 to 2π (full circle in radians)
# ─────────────────────────────────────────

# Generate heart points
heart_points = []

# t goes from 0 to 2*pi in small steps
# More steps = smoother heart shape
for t in np.linspace(0, 2 * np.pi, 300):

    # Heart formula (mathematical equations)
    x = 16 * (np.sin(t) ** 3)
    y = (13 * np.cos(t)
         - 5  * np.cos(2 * t)
         - 2  * np.cos(3 * t)
         -      np.cos(4 * t))

    # Scale up the heart (multiply by 15 to make it bigger)
    # Move to center of canvas (add 250, 280)
    # Flip y because screen y-axis is inverted
    px = int(x * 15 + 250)   # scale + move to center X
    py = int(-y * 15 + 280)  # flip + scale + move to center Y

    heart_points.append([px, py])

# Convert list to numpy array (required by OpenCV)
heart_points = np.array(heart_points, dtype=np.int32)

# ── Draw filled RED heart ──
cv2.fillPoly(canvas_heart,
             [heart_points],
             color=(0, 0, 255))   # RED in BGR

# ── Draw dark red border around heart ──
cv2.polylines(canvas_heart,
              [heart_points],
              isClosed=True,
              color=(0, 0, 150),  # dark red border
              thickness=2)

# ── Add text inside heart ──
cv2.putText(canvas_heart,
            'LOVE',
            (195, 285),                  # position
            cv2.FONT_HERSHEY_SIMPLEX,    # font style
            1.2,                         # font size
            (255, 255, 255),             # WHITE text
            3)                           # thickness

# Display
plt.figure(figsize=(6, 6))
plt.imshow(canvas_heart)
plt.title('❤️ Heart Shape', fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
````

## 🎨 Final: All Shapes on One Canvas

````python
# Create one big canvas
canvas_all = np.ones((600, 800, 3), dtype=np.uint8) * 255

# ── Line ──
cv2.line(canvas_all, (30, 50), (180, 50), (0, 0, 255), 3)
cv2.putText(canvas_all, 'Line', (75, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,200), 1)

# ── Rectangle ──
cv2.rectangle(canvas_all, (220, 20), (380, 100), (255, 0, 0), 3)
cv2.putText(canvas_all, 'Rectangle', (255, 15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (200,0,0), 1)

# ── Filled Box ──
cv2.rectangle(canvas_all, (420, 20), (560, 100), (0, 200, 0), -1)
cv2.putText(canvas_all, 'Filled Box', (430, 15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,150,0), 1)

# ── Circle ──
cv2.circle(canvas_all, (680, 60), 50, (0, 165, 255), 3)
cv2.putText(canvas_all, 'Circle', (655, 15), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,130,200), 1)

# ── Triangle ──
tri = np.array([[80, 150], [20, 280], [150, 280]], dtype=np.int32)
cv2.polylines(canvas_all, [tri], True, (128, 0, 128), 3)
cv2.putText(canvas_all, 'Triangle', (30, 145), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (100,0,100), 1)

# ── Pentagon ──
cx, cy, r = 300, 220, 80
penta = np.array([
    [int(cx + r * math.cos(math.radians(90 + i*72))),
     int(cy - r * math.sin(math.radians(90 + i*72)))]
    for i in range(5)], dtype=np.int32)
cv2.fillPoly(canvas_all, [penta], (0, 200, 200))
cv2.putText(canvas_all, 'Pentagon', (255, 145), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,150,150), 1)

# ── Hexagon ──
cx2, cy2, r2 = 530, 220, 75
hexa = np.array([
    [int(cx2 + r2 * math.cos(math.radians(i*60))),
     int(cy2 + r2 * math.sin(math.radians(i*60)))]
    for i in range(6)], dtype=np.int32)
cv2.fillPoly(canvas_all, [hexa], (0, 165, 255))
cv2.putText(canvas_all, 'Hexagon', (490, 145), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,130,200), 1)

# ── Heart ──
heart = []
for t in np.linspace(0, 2*np.pi, 300):
    x = 16 * (np.sin(t)**3)
    y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
    heart.append([int(x*10 + 680), int(-y*10 + 390)])
heart = np.array(heart, dtype=np.int32)
cv2.fillPoly(canvas_all, [heart], (0, 0, 255))
cv2.putText(canvas_all, 'Heart', (655, 290), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,200), 1)

# Title on canvas
cv2.putText(canvas_all, 'All Shapes - OpenCV Drawing',
            (200, 560), cv2.FONT_HERSHEY_SIMPLEX, 1, (50,50,50), 2)

# Display
plt.figure(figsize=(14, 10))
plt.imshow(canvas_all)
plt.title('🎨 All Shapes Together', fontsize=16, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.savefig('all_shapes.png', dpi=150)
plt.show()
````

## ✅ Summary — Shape Drawing Cheatsheet

| Shape | Function | Key Parameter |
|-------|----------|--------------|
| Line | `cv2.line()` | start, end points |
| Rectangle | `cv2.rectangle()` | top-left, bottom-right |
| Circle | `cv2.circle()` | center, radius |
| Triangle | `cv2.polylines()` | 3 corner points |
| Polygon | `cv2.fillPoly()` | n corner points |
| Heart | `cv2.fillPoly()` | math formula points |

> 💡 **Remember:**
> - Color is always **(B, G, R)** in OpenCV
> - `thickness = -1` → **fills** the shape
> - Use `np.array(..., dtype=np.int32)` for polygon points
